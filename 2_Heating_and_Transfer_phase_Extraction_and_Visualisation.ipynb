# Heating and Transfer phase Extraction and Visualisation

After obtaining informations about the data and the process, one can undestand the production cycle of one part. It is separated into 3 phases: heating phase, transfer phase and forging phase. We will not go into details about the proces, for better understanding of the data and process splitting, look again at the `met4fof-strath-afrc-testbed` notebooks.

Here we will only summarize the most important information required for understanding of these notebooks:
   - raw part is heated inside the induction coil,
   - pyrometer 1 (`TMP_Ind_U1 [°C]`) is used for monitoring of temperature during the heating phase,
   - heating starts and stops are determined by sensor `$U_GH_HEATON_1 (U25S0).1` which is used for bounding of heating phase,
   - forging period is period where the forging force is bigger than zero,
   - transfer period start is determined by `$U_GH_HEATON_1 (U25S0).1` and transfer period end is the same as forging period start (part is transfered from the induction coil to the forging box).
   

### Data import...

```python
data=[0]*81
for i in range(81):
    print("|"*(i+1),(i+1), end="\r")
    file_format="Scope"+str("{:04d}".format(i+1))+".csv"    
    data[i] = pd.read_csv(Path('Data')/'AFRC Radial Forge - Zenodoo Upload v3'/'Data'/'ScopeTraces'/file_format.format(i), header=0, encoding = 'unicode_escape')#, index_col=0)
    time.sleep(1)
```

### Extracting Heating and Transfer Periods....

The signal `$U_GH_HEATON_1 (U25S0).1` indicates when the induction heater switches on and off, and Force signal will be used for Transfer period. Transfer period is from the end of heating until the start of forging phase.

```python
font = {'family' : 'Times New Roman', 'weight' : 'normal', 'size'   : 20}
mpl.rcParams['figure.figsize'] = (20,10)
mpl.rc('font', **font)

plt.plot(merged_data['Time [s]'], merged_data['$U_GH_HEATON_1 (U25S0).1'])
plt.xlabel('Time [s]')
plt.ylabel('Heater On/Off');
```

```python
digital_sig_heating = merged_data['$U_GH_HEATON_1 (U25S0).1']>0
heating_diff = digital_sig_heating.astype('int').diff()

digital_sig_forge = merged_data['Force [kN]']>0
forge_diff = digital_sig_forge.astype('int').diff()
```

```python
heating_start=heating_diff[heating_diff==1].index.values
heating_stop=heating_diff[heating_diff==-1].index.values
heating_phases=pd.DataFrame({'heating_start_ind': heating_start[:-1], 'heating_stop_ind':heating_stop})
heating_phases_times =pd.DataFrame({'heating_start_time': merged_data['Time [s]'][heating_diff==1].values[:-1], 'heating_stop_time':merged_data['Time [s]'][heating_diff==-1].values})

transfer_phases=pd.DataFrame({'transfer_start_ind': heating_stop, 'transfer_stop_ind':forge_diff[forge_diff==1].index.values})
transfer_phases_times=pd.DataFrame({'transfer_start_time':merged_data['Time [s]'][heating_diff==-1].values, 'transfer_stop_time':merged_data['Time [s]'][forge_diff==1].values})
```

```python
print(heating_phases.head())
print(heating_phases_times.head())
print(transfer_phases.head())
print(transfer_phases_times.head())
```

The whole time series of temperature sensor will be plotted here, with marks for heating phases (starts and stops). By default, it will be zoomed to show 3 phases (3 production parts), but user can change the `plt.xlim()` to see more or less phases.

```python
plt.figure(figsize=(20,10))
plt.plot(merged_data['Time [s]'], merged_data['TMP_Ind_U1 [°C]'])    #Force [kN]
for index, row in heating_phases_times.iterrows():
    plt.axvline(row.heating_start_time, c='c', ls='--')
    plt.axvline(row.heating_stop_time, c='r', ls='--')
    
for index, row in transfer_phases_times.iterrows():
    plt.axvline(row.transfer_stop_time, c='g', ls='--')   
    
plt.xlabel('Time [s]', fontsize=20)
plt.ylabel('Part temperature during heating [°C]', fontsize=20)
plt.tick_params(labelsize=20)
plt.text(30,600, '<--Heating start \n         (part 1)', color='c', fontsize=20)
plt.text(140,500, '<--Heating end \n        (part 1)\n  Transfer start', color='r',  fontsize=20)
plt.text(176,350, '<--Transfer end \n       (part 1)', color='g',  fontsize=20)
plt.xlim([25,200]) # 3 cycles
plt.show()
```

```python
heating_sensors = set()
```

During the heating phase, the most relevant sensor is pyrometer 1 which is monitoring the temperature during heating. That's why we will extract most of the features from that sensor in heating phase. Sensor for position of part during the heating could also be useful. There is nominal position (setpoint) and actual.

```python
heating_sensors.update(['TMP_Ind_U1 [°C]', 'IP_ActPos [mm]', 'IP_NomPos'])
```

```python
plt.figure(figsize=(20,8))

plt.subplot(111)
for index, row in heating_phases.iterrows():
    y=merged_data[row.heating_start_ind:row.heating_stop_ind]
    x=merged_data[row.heating_start_ind:row.heating_stop_ind]
    plt.plot(x['Time [s]'].values-x['Time [s]'].values[0],y['TMP_Ind_U1 [°C]'].values)
plt.xlabel('Time [s]')
plt.ylabel('TMP_Ind_U1 [°C]')
plt.title('Temperature during heating for all parts')
```

`heating_data` is the data for all parts, but only for selected sensors. 

```python
heating_data=pd.DataFrame(merged_data[heating_sensors])
```

We will keep the time column and ID labels.

```python
heating_data['Timer Tick [ms]']=merged_data['Timer Tick [ms]']
heating_data['traceID']=merged_data['traceID'].values
heating_data['Time [s]']=merged_data['Time [s]'].values

heating_data.head()
#heating_data.tail()
```

Heating phase will be transformed into matrices for each part and stored in list `heating_phase`. List has lenght of 81 becaue of 81 parts. Tha same applies for the transfer period, which will be stored in list `transfer_phase`.

```python
heating_phase = []

for index, row in heating_phases.iterrows():
    part_heating_phase_df = heating_data[row.heating_start_ind:row.heating_stop_ind].copy()
    part_heating_phase_df['partID']=part_heating_phase_df.iloc[part_heating_phase_df.index==row.heating_start_ind]['traceID'].values[0]
    part_heating_phase_df['Time [s]']=part_heating_phase_df['Time [s]']-part_heating_phase_df['Time [s]'].iloc[0]
    heating_phase.append(part_heating_phase_df)
    part_heating_phase_df.set_index('Time [s]', inplace=True)

heating_phase_df=pd.concat(heating_phase)
heating_phase_df.set_index('partID', append=True, inplace=True)
```

```python
plt.figure(figsize=(20,8))

plt.subplot(111)
for index, row in transfer_phases.iterrows():
    y=merged_data[row.transfer_start_ind:row.transfer_stop_ind]
    x=merged_data[row.transfer_start_ind:row.transfer_stop_ind]
    plt.plot(x['Time [s]'].values-x['Time [s]'].values[0],y['TMP_Ind_U1 [°C]'].values)
plt.xlabel('Time [s]')
plt.ylabel('TMP_Ind_U1 [°C]')
plt.title('Temperature during transfer for all parts')
```

```python
transfer_phase = []

for index, row in transfer_phases.iterrows():
    part_transfer_phase_df = heating_data[row.transfer_start_ind:row.transfer_stop_ind].copy()
    part_transfer_phase_df['partID']=part_transfer_phase_df.iloc[part_transfer_phase_df.index==row.transfer_start_ind]['traceID'].values[0]
    part_transfer_phase_df['Time [s]']=part_transfer_phase_df['Time [s]']-part_transfer_phase_df['Time [s]'].iloc[0]
    transfer_phase.append(part_transfer_phase_df)
    part_transfer_phase_df.set_index('Time [s]', inplace=True)

transfer_phase_df=pd.concat(transfer_phase)
transfer_phase_df.set_index('partID', append=True, inplace=True)
```

If we look at the pyrometer 1 signal during the transfer phase, we will see that the part is under the pyrometer even after heating phase finishes. The other two lines are nominal and actual positioning of part inside heating coils. When nominal (setpoint) value changes, that means the part starts moving from the heating zone.

Part of the transfer phase until this point will be taken as relevant for calculating the **heat loss rate** which will be used as one feature. The other feeatures from heating phase will be: **maximum temperature after heating phase** (last temperature before heating turns off), last relevant measurement of temperature before the part is out of sensor reach (**last temperature from transfer phase**), and **duration of transfer** phase.

```python
plt.plot(transfer_phase[0]['TMP_Ind_U1 [°C]'], label='temperature pyr 1')
plt.plot(transfer_phase[0]['IP_ActPos [mm]'], label='actual position on heating axis')
plt.plot(transfer_phase[0]['IP_NomPos'],label='setpoint position on heating axis')


peaksTransfer, _ = find_peaks(transfer_phase[0]['IP_NomPos'].diff(), height=200, prominence=200, distance=70)
plt.axvline(transfer_phase[0].index[peaksTransfer], c='y', ls='--')
plt.legend()
peaksTransfer
```

This part of the transfer phase where the part is loosing heat will be approximated with linear function, and slope of that function is the heat loss rate.

```python
plt.plot(transfer_phase[0]['TMP_Ind_U1 [°C]'], label='temperature pyr 1')
plt.plot(transfer_phase[0]['IP_ActPos [mm]'], label='actual position on heating axis')
plt.plot(transfer_phase[0]['IP_NomPos'],label='setpoint position on heating axis')

peaksTransfer, _ = find_peaks(transfer_phase[0]['IP_NomPos'].diff(), height=200, prominence=200, distance=70)

plt.axvline(transfer_phase[0].index[peaksTransfer], c='y', ls='--')
plt.xlim(-1,15)

heat_loss_coeffs=np.polyfit(transfer_phase[0].index[:int(peaksTransfer)], transfer_phase[0]['TMP_Ind_U1 [°C]'].iloc[:int(peaksTransfer)], 1)
plt.plot(transfer_phase[0].index[:int(peaksTransfer)], transfer_phase[0].index[:int(peaksTransfer)]*heat_loss_coeffs[0]+heat_loss_coeffs[1], label="linear fit")
plt.legend()
```

Extracting mentioned features from transfer and heating phase.

```python
heating_transfer_temp_features = pd.DataFrame(index=pd.RangeIndex(1,len(data)+1))

heating_transfer_temp_features['Max Temp After Heating [°C]']=np.nan

heating_transfer_temp_features['Transfer Time [s]']=np.nan

heating_transfer_temp_features['Heat Loss Rate [°C/s]']=np.nan

heating_transfer_temp_features['Last Temp from Transfer Phase [°C]']=np.nan
```

```python
for idx, df in enumerate(heating_phase):
    heating_transfer_temp_features['Max Temp After Heating [°C]'][idx+1]=df['TMP_Ind_U1 [°C]'].iloc[-1]
    
for idx, df in enumerate(transfer_phase):
    heating_transfer_temp_features['Transfer Time [s]'][idx+1]=df.index[-1]-df.index[0]
    
    peaksTransfer, _ = find_peaks(df['IP_NomPos'].diff(), height=200, prominence=200, distance=70)
    heat_loss_coeffs=np.polyfit(df.index[:int(peaksTransfer)], df['TMP_Ind_U1 [°C]'].iloc[:int(peaksTransfer)], 1)
    heating_transfer_temp_features['Heat Loss Rate [°C/s]'][idx+1]=heat_loss_coeffs[0]
    heating_transfer_temp_features['Last Temp from Transfer Phase [°C]'][idx+1]=df['TMP_Ind_U1 [°C]'].iloc[int(peaksTransfer)]
```
