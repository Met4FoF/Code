```python

from uncertainties import ufloat
import numpy as np
from uncertainties import unumpy
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
x=np.arange(0,10,1e-5)
FSample=1001
Sampletimes=np.arange(0,10,1/FSample)
FSignal=1000
SampleClock=np.sin(FSample*np.pi*2*x)
SampleClockBin = np.where(SampleClock >= 0.0, 1, 0)
Signal=np.sin(FSignal*np.pi*2*x)
Samples=np.sin(FSignal*np.pi*2*Sampletimes)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle("ADC Calibration Signals")
ax1.set_ylabel("Sample Clock logic Level")
ax1.plot(x[:1000],SampleClockBin[:1000],label='Sample clock 1001 Hz')
ax2.plot(x[:1000],Signal[:1000],Label="Signal 1000 Hz",color='orange')
ax2.set_ylabel(r"ADC Voltage in A.U")
ax2.set_xlabel(r"Time in s")
fig.legend()
fig.show()
```

```python
fig, (ax1) = plt.subplots(1, 1)
fig.suptitle("ADC Calibration Signals")
ax1.set_ylabel("Sampled Signal")
ax1.plot(x[:2000],Signal[:2000],Label="Reconstructed signal")
ax1.plot(Sampletimes[:20],Samples[:20],label='Sampled Signal',marker='.',color='orange')
ax1.set_xlabel(r"Time in s")
fig.legend()
fig.show()
```

```python
fig, (ax1) = plt.subplots(1, 1)
fig.suptitle("ADC Calibration Signals")
ax1.set_ylabel("Sampled Signal")
ax1.plot(Sampletimes,Samples,label='Sampled Signal',color='orange')
ax1.set_xlabel(r"Time in s")
fig.legend()
fig.show()
```
