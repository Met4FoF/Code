```python

from uncertainties import ufloat
import numpy as np
from uncertainties import unumpy
import matplotlib.pyplot as plt
from scipy.stats import chi2
scalefactor=1.3
SMALL_SIZE = 8*scalefactor
MEDIUM_SIZE = 10*scalefactor
BIGGER_SIZE = 12*scalefactor

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
x=np.arange(0,10,1e-5)
FSample=999
Sampletimes=np.arange(0,10,1/FSample)
FSignal=1000
SampleClock=np.sin(FSample*np.pi*2*x)
SampleClockBin = np.where(SampleClock >= 0.0, 1, 0)
Signal=np.sin(FSignal*np.pi*2*x)
Samples=np.sin(FSignal*np.pi*2*Sampletimes)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle("ADC Cal. Signal")
ax1.set_ylabel("Sample Clock ")
ax1.plot(x[:1000],SampleClockBin[:1000],label='Sample clock '+str(FSample)+' Hz',color='red')
ax2.plot(x[:1000],Signal[:1000],Label="Signal "+str(FSignal)+" Hz",color='orange')
ax2.set_ylabel(r"ADC Voltage in A.U")
ax2.set_xlabel(r"Time in s")
fig.show()
fig.savefig("ADC_Cal_Trigg_AND_sig.svg")
```

```python
fig, (ax1) = plt.subplots(1, 1)
fig.suptitle("ADC Calibration Signal")
ax1.set_ylabel("Sampled Signal")
ax1.plot(x[:2000],Signal[:2000],Label="Reconstructed signal",color='orange')
ax1.plot(Sampletimes[:20],Samples[:20],label='Sampled signal',marker='.',c='C0')
ax1.set_xlabel(r"Time in s")
fig.legend(loc='upper center',bbox_to_anchor=(0.7, 0.3))
fig.show()
fig.savefig("ADC_reconstructed_Signal.svg")
```

```python
fig, (ax1) = plt.subplots(1, 1)
fig.suptitle("ADC Calibration Signal")
ax1.set_ylabel("Sampled Signal")
ax1.plot(Sampletimes,Samples,label='Sampled Signal',c='C0')
ax1.set_xlabel(r"Time in s")
fig.show()
fig.savefig("ADC_Sampled_Signal.svg")
```
